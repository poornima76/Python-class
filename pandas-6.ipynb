```python
def convert_areatype(s):
    if s == 'Super built-up Area':
        return 1
    elif s == 'Built-up  Area':
        return 2
    elif s == 'Plot  Area':
        return 3
    elif s =='Carpet  Area':
        return 4
```

```python
df['area_type'] = df['area_type'].apply(convert_areatype)
```
