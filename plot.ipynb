```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
companies =['fb', 'ig','viber','twitter']
```

```python
revenue = [2000,3000,4000,3500]
```

```python
plt.bar(companies, revenue, color=['r','b','g'])
```

```python
plt.bar(companies, revenue, color=['g','b'])
```

```python
countries = ['USA', 'Nepal', 'Bangladesh', 'Bhutan']
gold_medal = np.array([7, 5, 3, 2])
silver_medal = np.array([3, 4, 2, 5])
bronze_medal = np.array([4, 2, 7, 8])
```

```python
plt.bar(countries, gold_medal, label='gold')
plt.bar(countries, silver_medal, bottom = gold_medal, label='silver')
plt.bar(countries, bronze_medal, bottom = gold_medal + silver_medal, label='bronze')
plt.grid(axis='y')
plt.legend()
```

### if the bars are very conjested we use horizontal bar graph

```python
plt.barh(companies, revenue, color=['g','b'])
```

```python
expenditures = ['momo', 'rent', 'phone', 'chiya']
paisa = [2500, 4000, 200, 5000]
```

```python
plt.pie(paisa, labels = expenditures)
```

```python
plt.pie(paisa, labels = expenditures, autopct='%0.2f%%', explode=[0.3,0,0,0])
```
