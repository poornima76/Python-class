```python
df = pd.read_csv('titanic_test.csv')
```

```python
df.Age.fillna(df.Age.mean(), inplace=True)
```

df['Date']= pd.to_datetime(df['Date']) for date

```python
df.Age.loc[0] 
```

```python
df.loc[0, 'Age'] = 105
```

```python
for index in df.index:
    print(df.loc[index, 'Age'])
```

```python
for index in df.index:
    if df.loc[index, 'Age'] > 90:
        df.loc[index, 'Age'] = 90
```

```python
df.Age.max()
```

### since Age cannot be 105 so we cleaned the data by replacing all the bad value which is greater than 90 to 90.
