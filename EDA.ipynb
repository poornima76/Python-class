```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('titanic_test.csv')
```

```python
df.columns
```

```python
age_vs_survived = df[['Sex', 'Survived']]
```

```python
male = age_vs_survived[age_vs_survived.Sex == 'male']
```

```python
sum(male.Survived == 0)
```

```python
sum(male.Survived == 1)
```

```python
female = age_vs_survived[age_vs_survived.Sex == 'female']
```

```python
sum(female.Survived == 0)
```

```python
sum(female.Survived == 1)
```

```python
plt.bar(['Male', 'Female'],[sum(male.Survived == 0), sum(female.Survived == 0)], label='Died')
plt.bar(['Male', 'Female'],[sum(male.Survived == 1), sum(female.Survived == 1)], label= 'Survived' ,bottom=[sum(male.Survived == 0), sum(female.Survived == 0)])
plt.legend()
```
